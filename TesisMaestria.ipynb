```python
import pandas as pd
import random
import string

# Generar lista de nombres y apellidos
nombres = ["Ana", "Luis", "María", "Carlos", "Juana", "Pedro", "Lucía", "Jorge", "Marta", "Raúl", "Elena", "Miguel", "Laura", "José", "Carmen", "Francisco", "Sara", "Andrés", "Rosa", "Enrique"]
apellidos = ["Mamani", "Chavez", "Huanca", "Gutierrez", "Caceres", "Torres", "Quispe", "Morales", "Choque", "Choquemamani", "Benitez", "Aguilar", "Apaza", "Huallpa", "Condori", "Vilca", "Huamani", "Flores", "Tupa"]

# Definir notas con mayor concentración en [8,10,12]
notas = [2, 4, 6, 8, 10, 12, 14]
pesos = [0.05, 0.05, 0.05, 0.25, 0.30, 0.25, 0.05]

# Prefijos de los códigos de matrícula
prefijos = ["2214", "2228", "2220", "2216"]

# Generar datos
data = []
for i in range(358):
    codigo_matricula = random.choice(prefijos) + ''.join(random.choices(string.digits, k=2))
    nombre = random.choice(nombres)
    apellido1 = random.choice(apellidos)
    apellido2 = random.choice(apellidos)
    sexo = "F" if random.random() < 0.65 else "M"
    nota1 = random.choices(notas, weights=pesos, k=1)[0]
    nota2 = random.choices(notas, weights=pesos, k=1)[0]
    promedio = round((nota1 + nota2) / 2, 2)
    data.append([codigo_matricula, nombre, apellido1, apellido2, sexo, nota1, nota2, promedio])

# Crear DataFrame
df = pd.DataFrame(data, columns=["Codigo Matricula", "Nombres", "Apellido1", "Apellido2", "Sexo", "Nota1", "Nota2", "Promedio"])

# Guardar en un archivo Excel
df.to_excel("registros_358.xlsx", index=False)
print("Archivo Excel generado con éxito.")
```

```python
import pandas as pd
import random

# Generar lista de nombres
nombres = ["Geanella","Jordi","Luanda","Brayan","Brandol","Jefferson","Ibeth","Alexis","Estrella","Criss","Franz","Katherine","Jordan","Gilber","Vicente","Vicenta","Dayanna","Yessica","Alba","Jeniffer","Angeles","Williams","Ruth","Yanina","Yenny","Betsi","Daisha","Alely","Astrid", "Pedro", "Lucía", "Jorge", "Marta", "Raúl", "Elena", "Miguel", "Laura", "José", "Carmen", "Francisco", "Sara", "Andrés", "Rosa", "Enrique"]

# Generar nombres compuestos
data = []
for i in range(50):
    nombre1 = random.choice(nombres)
    nombre2 = random.choice(nombres)
    while nombre2 == nombre1:
        nombre2 = random.choice(nombres)
    nombre_compuesto = f"{nombre1} {nombre2}"
    data.append([nombre_compuesto])

# Crear DataFrame
df = pd.DataFrame(data, columns=["Nombres Compuestos"])

# Guardar en un archivo Excel
df.to_excel("nombres_compuestos.xlsx", index=False)
print("Archivo Excel generado con éxito.")
```

```python
import pandas as pd
import random

# Definir notas con menor cantidad de casos para 12 y 18
notas = [12, 14, 16, 18]
pesos = [0.15, 0.35, 0.35, 0.15]  # Pesos ajustados para que 12 y 18 sean menos frecuentes

# Generar datos
data = []
for i in range(93):
    nota1 = random.choices(notas, weights=pesos, k=1)[0]
    nota2 = random.choices(notas, weights=pesos, k=1)[0]
    promedio = round((nota1 + nota2) / 2, 2)
    data.append([nota1, nota2, promedio])

# Crear DataFrame
df = pd.DataFrame(data, columns=["Nota1", "Nota2", "Promedio"])

# Guardar en un archivo Excel
df.to_excel("notas_93_registros.xlsx", index=False)
print("Archivo Excel generado con éxito.")
```
