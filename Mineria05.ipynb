#Ejercicios para practicar con la función loc
A continuación, se presentan algunos ejercicios para practicar con la función loc en Pandas:

1. Seleccione la fila con etiqueta 'persona4' y la columna 'Ciudad' del DataFrame df.

2. Seleccione las filas con etiquetas 'persona2' y 'persona3' y todas las columnas del DataFrame df.

3. Cree un DataFrame llamado df2 con los siguientes datos:
                    {'A': [1, 4, 7],
                    'B': [2, 5, 8],
                    'C': [3, 6, 9]})

4. Utilice la función loc para seleccionar las filas con etiquetas 0 y 2 y todas las columnas del DataFrame df2.




```python
# Importar la biblioteca pandas
import pandas as pd

# Crear un DataFrame df
data = {'Nombre': ['persona1', 'persona2', 'persona3', 'persona4'],
        'Ciudad': ['Ciudad1', 'Ciudad2', 'Ciudad3', 'Ciudad4'],
        'Edad': [25, 30, 35, 40]}

df = pd.DataFrame(data)

# 1. Seleccionar la fila con etiqueta 'persona4' y la columna 'Ciudad'
seleccion_1 = df.loc[df['Nombre'] == 'persona4', 'Ciudad']
print("1. Fila con etiqueta 'persona4' y columna 'Ciudad':\n", seleccion_1)

# 2. Seleccionar las filas con etiquetas 'persona2' y 'persona3' y todas las columnas
seleccion_2 = df.loc[df['Nombre'].isin(['persona2', 'persona3']), :]
print("\n2. Filas con etiquetas 'persona2' y 'persona3' y todas las columnas:\n", seleccion_2)

# 3. Crear un DataFrame df2
data2 = {'A': [1, 4, 7],
         'B': [2, 5, 8],
         'C': [3, 6, 9]}

df2 = pd.DataFrame(data2)
print("\n3. DataFrame df2:\n", df2)

# 4. Utilizar la función loc para seleccionar las filas con etiquetas 0 y 2 y todas las columnas de df2
seleccion_4 = df2.loc[[0, 2], :]
print("\n4. Filas con etiquetas 0 y 2 y todas las columnas de df2:\n", seleccion_4)
```
