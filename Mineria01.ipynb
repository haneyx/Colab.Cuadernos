## Ejercicio 1
Dise침e un programa que determine la Superficie (S) y el Per칤metro (P) de un c칤rculo del que se conoce su radio (r) . Considere las siguientes f칩rmulas: S=2洧랢洧 y P=洧랢 洧

```python
import math

# Solicitar radio
radio = float(input("Ingresa el valor del radio del c칤rculo: "))

# Superficie
S = 2 * math.pi * radio

# Calcular el per칤metro
P = math.pi * radio

# Imprimir los resultados
print(f"La superficie del c칤rculo es: {S}")
print(f"El per칤metro del c칤rculo es: {P}")
```

## Ejercicio 2

Escribir un PROGRAMA EN PYTHON de un programa que permita leer el peso de una persona, calcule el IMC y posteriormente imprimirla.
* F칩rmula: IMC = Peso/(Estatura x Estatura)

```python
# Ingrese peso
peso = float(input("Ingresa tu peso: "))

# Ingrese estatura
estatura = float(input("Ingresa tu estatura en metros: "))

# Calcular el IMC
imc = peso / (estatura ** 2)

# Imprimir el resultado del IMC
print(f"Tu 칈ndice de Masa Corporal (IMC) es: {imc:.2f}")
```

## Ejercicio 3
Determinar la frecuencia cardiaca. El usuario debe proporcionarte su edad y su peso.
* Formula: freq = 210 - 0.5 x edad - (0.01 x peso + 4)


```python
# Ingrese peso y edad
edad = int(input("Ingresa tu edad: "))
peso = float(input("Ingresa tu peso en kilogramos: "))

# Calcular la frecuencia card칤aca
freq = 210 - 0.5 * edad - (0.01 * peso + 4)

# Imprimir el resultado
print(f"Tu frecuencia card칤aca m치xima es: {freq:.2f} latidos por minuto")
```

## Ejercicio 4

Hacer un formulario para registrar sus nombres, apellidos, fecha de nacimiento, su nota parcial y nota final. Una vez registrado sus datos mostrar en pantalla que se ha registrado con 칠xitos sus datos y la edad del participante y su promedio de notas.


```python
import datetime

# Solicitar al usuario que ingrese los datos
nombres = input("Ingresa tus nombres: ")
apellidos = input("Ingresa tus apellidos: ")
fecha_nacimiento = input("Ingresa tu fecha de nacimiento (formato: DD-MM-YYYY): ")
nota_parcial = float(input("Ingresa tu nota parcial: "))
nota_final = float(input("Ingresa tu nota final: "))

# Calcular la edad a partir de la fecha de nacimiento
fecha_nacimiento = datetime.datetime.strptime(fecha_nacimiento, '%d-%m-%Y')
hoy = datetime.datetime.now()
edad = hoy.year - fecha_nacimiento.year - ((hoy.month, hoy.day) < (fecha_nacimiento.month, fecha_nacimiento.day))

# Calcular el promedio de notas
promedio_notas = (nota_parcial + nota_final) / 2

# Mostrar los datos registrados exitosamente
print("Datos registrados exitosamente!")
print(f"Nombre: {nombres} {apellidos}")
print(f"Edad: {edad} a침os")
print(f"Promedio de notas: {promedio_notas:.2f}")
```
