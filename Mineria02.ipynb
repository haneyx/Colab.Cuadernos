## Ejercicio 1 - CADENAS
* Dado una cadena de con el siguiente formato de hora: hhmmss (182450)
* mostrar el siguiente mensaje : “Son las 18h:24min:50seg”

```python
# Solicitar al usuario que ingrese el formato de hora
hora_str = input("Ingresa un formato de hora en hhmmss: ")

# Validar que la entrada tenga al menos 6 caracteres
if len(hora_str) != 6:
    print("El formato de hora debe tener 6 caracteres (hhmmss).")
else:
    # Extraer las horas, minutos y segundos
    horas = int(hora_str[0:2])
    minutos = int(hora_str[2:4])
    segundos = int(hora_str[4:6])

    # Mostrar la hora, los minutos y los segundos
    print(f"Son las {horas} horas {minutos} minutos {segundos} segundos.")
```

## Ejercicio 2 - CADENAS

DIFERENCIA DE HORAS
* HORA INICIAL: 123050
* HORA FINAL=184530
* MOSTRAR EL MENSAJE "La diferencia es 18h:24min:50seg"

```python
from datetime import datetime

# Hora inicial y hora final en formato de cadena
hora_inicial_str = "123050"
hora_final_str = "184530"

# Convierte las cadenas en objetos de tiempo
hora_inicial = datetime.strptime(hora_inicial_str, "%H%M%S")
hora_final = datetime.strptime(hora_final_str, "%H%M%S")

# Calcula la diferencia entre las dos horas
diferencia = hora_final - hora_inicial

# Imprime la diferencia en el formato deseado
print(f"La diferencia es {diferencia.seconds // 3600}h:{(diferencia.seconds // 60) % 60}min:{diferencia.seconds % 60}seg")
```

## Ejercicio 4 - CONDICIONALES

Una tienda vende un producto a precios unitarios que dependen de la cantidad de unidades adquiridas de acuerdo a la tabla adjunta:

* Adicionalmente, si el cliente adquiere más de 50 unidades la tienda le descuenta el 15% del importe de la compra; en caso contrario, sólo le descuenta el 5% del importe compra.
* Diseñe un programa que determine el importe de la compra, el importe del descuento y el importe a pagar por la compra de cierta cantidad de unidades del producto.

```python
# Precios unitarios por cantidad de unidades adquiridas
precio_unitario = {
    1: 10.0,
    10: 9.0,
    20: 8.0,
    50: 7.0,
    51: 6.5  # Precio con descuento del 15%
}

# Solicita la cantidad de unidades que el cliente desea comprar
cantidad = int(input("Ingrese la cantidad de unidades a comprar: "))

# Determina el precio unitario según la cantidad
precio = None

for cantidad_limite, precio_unitario_actual in precio_unitario.items():
    if cantidad <= cantidad_limite:
        precio = precio_unitario_actual
        break

# Calcula el importe de la compra
importe_compra = cantidad * precio

# Calcula el descuento
if cantidad > 50:
    descuento = 0.15 * importe_compra
else:
    descuento = 0.05 * importe_compra

# Calcula el importe a pagar
importe_a_pagar = importe_compra - descuento

# Imprime los resultados
print(f"Importe de la compra: ${importe_compra:.2f}")
print(f"Importe del descuento: ${descuento:.2f}")
print(f"Importe a pagar: ${importe_a_pagar:.2f}")
```

## Ejercicio 5 - Bucles

Calcular el sueldo promedio:

1. Ingresa la cantidad de sueldo a promediar
2. Ingresar los sueldos
3. Calcular el promedio de los sueldo.

```python
# Solicita la cantidad de sueldos a promediar
n = int(input("Ingrese la cantidad de sueldos a promediar: "))

# Inicializa una lista para almacenar los sueldos
sueldos = []

# Solicita los sueldos y los almacena en la lista
for i in range(n):
    sueldo = float(input(f"Ingrese el sueldo {i + 1}: "))
    sueldos.append(sueldo)

# Calcula el promedio de los sueldos
if n > 0:
    promedio = sum(sueldos) / n
    print(f"El sueldo promedio es: {promedio:.2f}")
else:
    print("No se ingresaron sueldos para calcular el promedio.")
```

## Ejercicio 6 - BUCLES + CONDICIONALES

Escribir un programa que almacene la cadena de caracteres contraseña en una variable, pregunte al usuario por la contraseña hasta que introduzca la contraseña correcta.

```python
# Establece la contraseña correcta
contrasena_correcta = "mi_contraseña_secreta"

# Solicita al usuario que ingrese la contraseña
while True:
    contrasena_ingresada = input("Ingrese la contraseña: ")

    # Verifica si la contraseña ingresada es correcta
    if contrasena_ingresada == contrasena_correcta:
        print("¡Contraseña correcta! Acceso concedido.")
        break
    else:
        print("Contraseña incorrecta. Inténtalo de nuevo.")
```
